**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#transpose the dataframe, then check its updated row indexes
properties_trans = properties.T
print(properties_trans.index)
properties_trans.head()
```

```python
#reset the index, then check the updated row indexes
properties_trans = properties_trans.reset_index()
print(properties_trans.index)
properties_trans.head()
```

```python
#assign the row of dates to be column headings
properties_trans.columns = properties_trans.iloc[0]
properties_trans.head()
```

```python
#delete the redundant row of dates
properties_trans = properties_trans.drop(0)
properties_trans.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#Rename the first two columns with more descriptive names
properties_renamed = properties_trans.rename(columns = {"Unnamed: 0":"Borough", pd.NaT:"ID"})
properties_renamed.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Melt the data
properties_melted = properties_renamed.melt(id_vars = ['Borough', 'ID'], var_name = 'Date', value_name = 'Avg_price')
properties_melted.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#check the data types
properties_melted.info()
```

```python
#update the dtype of the Avg_price column, then recheck the data
properties_melted['Avg_price'] = properties_melted['Avg_price'].astype(float)
properties_melted.info()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#checking for unique values in the Boroughs column
properties_melted['Borough'].unique()
```

```python
#making lists of unusual column titles
unwanted_locations = ['Inner London', 'Outer London', 'NORTH EAST',
       'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS',
       'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST',
       'SOUTH WEST', 'England']
unnamed_boroughs = ['Unnamed: 34', 'Unnamed: 37', 'Unnamed: 47']

#exploring the unnamed_boroughs data
unnamed_data = properties_melted[properties_melted['Borough'].isin(unnamed_boroughs)]

print(unnamed_data.info())
unnamed_data.head()
```

```python
#removing the "Unnamed" rows, because they contain no valuable information. Checking to make sure they were removed properly
properties_cleaned = properties_melted.dropna()
properties_cleaned.info()
```

```python
#Removing the rows containing unwanted locations, then verifying that they were removed properly and only the wanted data remains
properties_cleaned = properties_cleaned[~ properties_cleaned['Borough'].isin(unwanted_locations)]
print(properties_cleaned.info())
properties_cleaned['Borough'].value_counts()
```

```python
#set the dataframe to the final variable df, then take a summary look at the clean data
df = properties_cleaned.sort_values(['Borough', 'Date'])
print(df.head())
df.groupby('Borough').describe()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#group df so that the upcoming function runs more quickly. Note that the borough and year are now in the index
dfg = df.groupby(['Borough', 'Year']).mean()
#type(dfg.loc[('Barking & Dagenham', 1995)]['Avg_price'])
dfg
```

```python
#defining the function
#I went off script here and calculated a % increase in price rather than a ratio.
#I prefer this because it's more tangible to me, and I feel I'll be able to present the findings more effectively.
#If it's a problem, please let me know, and I'll happily redo it.
def percent_increase(df_func, borough, yr_start, yr_end):
    start_price = df_func.loc[(borough, yr_start)]['Avg_price']
    end_price = df_func.loc[(borough, yr_end)]['Avg_price']
    #calculate and return the ratio
    return 100 * (end_price - start_price)/start_price

#test the function
percent_increase(dfg, 'Barking & Dagenham', 1998, 2018)
```

```python
#create an empty dictionary to hold each borough's price ratio
increases = {}

#iterate through the borough and calculate each price ratio, adding the borough name and ratio to the dictionary
for b in df['Borough']:
    increases[b] = [percent_increase(dfg, b, 1998, 2018)]

print(increases)
```

```python
#turn the dictionary into a dataframe
incr_df = pd.DataFrame.from_dict(increases)

incr_df
```

```python
#transform ratios_df, reset the index, rename the columns, and sort by the ratios descending

final_df = incr_df.T.reset_index().rename(columns = {'index':'Borough', 0:'Percent_incr'}).sort_values('Percent_incr', ascending = False)
final_df.head()
```

```python
#plot the data, based on the new, final dataframe
final_df.plot(kind = 'bar', x = 'Borough', y = 'Percent_incr')
plt.show()
```
